```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.cluster.hierarchy as sch

# Load the dataset
dataset = pd.read_csv("https://raw.githubusercontent.com/salemprakash/EDA/main/Data/student-mat.csv")
print(dataset.head(4))
# Extract only the final grades columns for clustering
X = dataset[['G1', 'G2', 'G3']].values

# Plot distribution of each grade
plt.figure(figsize=(8,5))
plt.title("G1 Grade Distribution", fontsize=16)
plt.xlabel("Grade G1", fontsize=14)
plt.grid(True)
plt.hist(dataset['G1'], color='blue', edgecolor='k')
plt.show()

plt.figure(figsize=(8,5))
plt.title("G2 Grade Distribution", fontsize=16)
plt.xlabel("Grade G2", fontsize=14)
plt.grid(True)
plt.hist(dataset['G2'], color='purple', edgecolor='k')
plt.show()

plt.figure(figsize=(8,5))
plt.title("G3 Grade Distribution", fontsize=16)
plt.xlabel("Grade G3", fontsize=14)
plt.grid(True)
plt.hist(dataset['G3'], color='green', edgecolor='k')
plt.show()

# Perform hierarchical clustering with a dendrogram
plt.figure(figsize=(15,6))
plt.title('Dendrogram for Final Grades')
plt.xlabel('Students')
plt.ylabel('Euclidean Distances')
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.show()

# Redraw with a cutoff line to suggest a cluster number
plt.figure(figsize=(15,6))
plt.title('Dendrogram with Cutoff Line')
plt.xlabel('Students')
plt.ylabel('Euclidean Distances')
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.hlines(y=50, xmin=0, xmax=2000, colors='r', linestyles='--', lw=2)
plt.text(x=800, y=60, s='Suggested Cluster Cutoff', fontsize=14, color='red')
plt.show()
```
