```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/salemprakash/EDA/main/Data/student-mat.csv')
print(df.head(3))
print(df.tail(3))
df
```

```python
df.dtypes
```

```python
print(df.describe())
pd.isna(df).any()
```

```python
df.isnull().sum()
```

```python
df.school.value_counts().nlargest(30).plot(kind='bar',figsize=(10,5))
plt.title("Student ratio")
plt.ylabel("Student Count")
plt.xlabel("School")
```

```python
sns.FacetGrid(df).map(sns.distplot,"age").add_legend()
```

```python
sns.FacetGrid(df).map(sns.distplot,"absences").add_legend()
```

```python
sns.boxplot(x='studytime',data=df)
plt.show()
```

```python
plt.scatter(df['age'],df['absences'])
plt.title('Age vs Absences')
plt.xlabel('Age')
plt.ylabel('Absences')
```

```python
sns.boxplot(x='sex',y='absences',data=df)
plt.show()
```

```python
sns.boxplot(x='sex',y='age',data=df)
plt.show()
```

```python
sns.pairplot(df,vars=['age','absences'],hue='sex')
plt.show()
```

```python
sns.pairplot(df,vars=['age','absences'],kind='reg')
plt.show()
```

```python
from scipy import stats
corr = stats.pearsonr(df['age'],df['absences'])
print("p-value:\t",corr[1])
print("correlation:\t",corr[0])
```

```python
correlation=df.corr(method='pearson',numeric_only=True)
correlation
```

```python
sns.heatmap(correlation,xticklabels=correlation.columns,yticklabels=correlation.columns)
plt.show()
```
